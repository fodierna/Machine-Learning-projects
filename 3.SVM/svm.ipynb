# Training
We learn 17 SV (8 for green class, 9 for red class). There aren't margin errors and the SV are in the exptected position.

```python
import random
#SVC with linear kernel
classifier = SVC(kernel = "linear")

#training classifier on data
classifier.fit(X_train, y_train)

#get support vectores learnt
SV = classifier.support_vectors_
y_SV = ["blue" if x + y > 0.3 else "orange"  for [x,y] in SV]
dual_weigth = classifier.dual_coef_
#print("Number of SV learnt: " + str(classifier.n_support_))
#print("Associated dual weights: \n" + str(classifier.dual_coef_))

#plotting training examples
plt.scatter(X_train[:,0], X_train[:,1], c=y_train)
#Plotting SV learnt
plt.scatter(SV[:,0], SV[:,1], c=y_SV)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = classifier.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
# plot support vectors
ax.scatter(SV[:, 0], SV[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
plt.show()
```
